```python
import numpy as np
import xarray as xr
```

```python
bands = np.random.rand(400_000, 55, 10).astype(np.float16)
doys = np.random.randint(1, 366, size=(400_000, 55)).astype(np.uint16)

np.savez_compressed("testing_xarray/s2_only.npz", bands=bands, doys=doys)
```

```python
!du -h testing_xarray/s2_only.npz
```

```python
# Carregar os dados
data = np.load("testing_xarray/s2_only.npz")
bands = data["bands"]  # shape: (400_000, 55, 10)
doys = data["doys"]     # shape: (400_000, 55)

n_samples, n_timestamps, n_bands = bands.shape

# Criar o Dataset
ds = xr.Dataset(
    data_vars={
        "reflectance": (("sample", "time", "band"), bands),
        "doy": (("sample", "time"), doys)
    },
    coords={
        "sample": np.arange(n_samples),
        "time": np.arange(n_timestamps),
        "band": ["blue", "green", "red", "re1", "re2", "re3", "nir", "re4", "swir1", "swir2"]
    }
)

zarr_path = "testing_xarray/s2_only.zarr"

ds.to_zarr(zarr_path, mode="w", consolidated=True)
```

```python
s2_bands = np.random.rand(400_000, 55, 10).astype(np.float16)
s2_doys = np.random.randint(1, 366, size=(400_000, 55)).astype(np.uint16)

l8_bands = np.random.rand(400_000, 30, 6).astype(np.float16)
l8_doys = np.random.randint(1, 366, size=(400_000, 30)).astype(np.uint16)

np.savez_compressed("testing_xarray/s2_l8_only.npz",
                    s2_bands=s2_bands, s2_doys=s2_doys,
                    l8_bands=l8_bands, l8_doys=l8_doys)
```

```python
!du -h testing_xarray/s2_l8_only.npz
```

```python
# Carregar dados
data = np.load("testing_xarray/s2_l8_only.npz")
s2_bands = data["s2_bands"]  # (400_000, 55, 10)
s2_doys = data["s2_doys"]    # (400_000, 55)
l8_bands = data["l8_bands"]  # (400_000, 30, 6)
l8_doys = data["l8_doys"]    # (400_000, 30)

n_samples = s2_bands.shape[0]

# Criar Dataset
ds = xr.Dataset(
    data_vars={
        "s2_reflectance": (["sample", "s2_time", "s2_band"], s2_bands),
        "s2_doy": (["sample", "s2_time"], s2_doys),
        "l8_reflectance": (["sample", "l8_time", "l8_band"], l8_bands),
        "l8_doy": (["sample", "l8_time"], l8_doys),
    },
    coords={
        "sample": np.arange(n_samples),
        "s2_time": np.arange(s2_bands.shape[1]),
        "s2_band": ["blue", "green", "red", "re1", "re2", "re3", "nir", "re4", "swir1", "swir2"],
        "l8_time": np.arange(l8_bands.shape[1]),
        "l8_band": ["blue", "green", "red", "nir", "swir1", "swir2"]
    }
)

zarr_path = "testing_xarray/s2_l8_only.zarr"

ds.to_zarr(zarr_path, mode="w", consolidated=True)
```
