```python
import cryptoservice
import os
from pprint import pprint
from dotenv import load_dotenv

load_dotenv()

api_key = os.getenv("BINANCE_API_KEY")
api_secret = os.getenv("BINANCE_API_SECRET")
if api_key is None or api_secret is None:
    raise ValueError("API key or secret is not set")
binance_client = cryptoservice.BinanceClientFactory.create_client(api_key, api_secret)

# data = binance_client.get_all_tickers()
futures_info = binance_client.futures_exchange_info()
perpetual_symbols = [
    symbol["symbol"]
    for symbol in futures_info["symbols"]
    if symbol["contractType"] == "PERPETUAL" and symbol["status"] == "TRADING"
]

pprint(perpetual_symbols)
```
